### Added new column "Dominant_Activity' (which age group is largest ?)

```python
# 1. Define the mapping for readable names
activity_map = {
    'age_0_5': 'New Births',
    'age_5_17': 'School Updates',
    'age_18_greater': 'Adult Corrections'
}

# 2. Find the column with the max value and map it
# axis=1 checks across columns (horizontally)
df['Dominant_Activity'] = df[['age_0_5', 'age_5_17', 'age_18_greater']].idxmax(axis=1).map(activity_map)
```

#### Replace '&' with 'and'. then converted everything to Title case

```python
# 1. Replace ALL '&' with 'and' in the entire column
df['state'] = df['state'].str.replace('&', 'and', regex=False)

# 2. Convert everything to "Title Case" (e.g., "AND" becomes "And")
# This ensures "Andaman And Nicobar" matches "Andaman and Nicobar"
df['state'] = df['state'].str.title()

# 3. Clean up accidental spaces at the start/end
df['state'] = df['state'].str.strip()

# 4. Remove that weird "100000" row (Data Cleaning)
# We filter to keep only rows where the state is NOT '100000'
df = df[df['state'] != '100000']

# --- NOW Run Grouping Again ---
state_grouped = df.groupby('state')[['age_0_5', 'age_5_17', 'age_18_greater']].sum()

print("Final Cleaned Grouping:")
print(state_grouped)
```

```python
# 1. Count Exact Duplicates (Complete row matches)
exact_dupes = df.duplicated().sum()
print(f"Exact Duplicates: {exact_dupes}")

# 2. Count Logical Duplicates (Same Date + Same Pincode)
# This finds if a specific location appears twice on the same day
logical_dupes = df.duplicated(subset=['date', 'pincode']).sum()
print(f"Same Date + Location Duplicates: {logical_dupes}")

# --- THE FIX ---
if exact_dupes > 0 or logical_dupes > 0:
    print("Removing duplicates...")
    # keep='last' keeps the most recent entry if duplicates exist
    df = df.drop_duplicates(subset=['date', 'pincode'], keep='last')
    print("Duplicates removed!")
else:
    print("No duplicates found. Dataset is clean.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# --- ASSUMPTION: 'df' is your clean dataframe from the previous step ---
# If you need to reload it, make sure to run the cleaning steps (duplicates/renaming) first.

# ==========================================
# INSIGHT 1: Normalized "State DNA" (100% Bar)
# ==========================================
# 1. Calculate the percentage of each activity within each state
state_activity = df.groupby(['state', 'Dominant_Activity']).size().unstack(fill_value=0)
state_pct = state_activity.div(state_activity.sum(axis=1), axis=0) * 100

# 2. Plot
ax = state_pct.plot(kind='bar', stacked=True, figsize=(14, 7), colormap='viridis', width=0.8)

plt.title('State Strategy DNA: Relative Focus of Each State (0-100%)', fontsize=16)
plt.ylabel('Percentage of Districts (%)', fontsize=12)
plt.xlabel('State', fontsize=12)
plt.legend(title='Activity Type', bbox_to_anchor=(1.0, 1.0))
plt.axhline(y=50, color='white', linestyle='--', alpha=0.3) # 50% marker
plt.tight_layout()
plt.show()


# ==========================================
# INSIGHT 2: The "Power Players" (Pareto Analysis)
# ==========================================
# Group by District to see who contributes the most volume
district_stats = df.groupby(['state', 'district'])['Total_count'].sum().reset_index()
district_stats = district_stats.sort_values(by='Total_count', ascending=False)

# Calculate cumulative percentage
total_work = district_stats['Total_count'].sum()
district_stats['Cumulative_Pct'] = (district_stats['Total_count'].cumsum() / total_work) * 100

print("\n--- TOP 5 'VIP' DISTRICTS (Highest Volume) ---")
print(district_stats.head(5))

# Check how many districts do 80% of the work
top_80_percent_count = district_stats[district_stats['Cumulative_Pct'] <= 80].shape[0]
total_districts = district_stats.shape[0]
print(f"\n--- PARETO INSIGHT ---")
print(f"{top_80_percent_count} districts (out of {total_districts}) contribute 80% of the total work.")


# ==========================================
# INSIGHT 3: Outlier Detection (Adult Activity Spikes)
# ==========================================
# We define an outlier as any row where 'age_18_greater' is exceptionally high
# (e.g., more than mean + 3 * standard_deviation)
threshold = df['age_18_greater'].mean() + (3 * df['age_18_greater'].std())
outliers = df[df['age_18_greater'] > threshold].sort_values(by='age_18_greater', ascending=False)

print(f"\n--- OUTLIER ALERT: Unusual Adult Activity (Threshold: > {int(threshold)}) ---")
if not outliers.empty:
    print(f"Found {len(outliers)} suspicious/high-activity locations.")
    print(outliers[['date', 'state', 'district', 'pincode', 'age_18_greater']].head(5))
else:
    print("No extreme anomalies found in adult data.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# --- STEP 1: PREPARE THE DATA ---
# Ensure date is in datetime format
df['date'] = pd.to_datetime(df['date'], format='%d-%m-%Y')

# Group data by Month to make the chart readable
# We resample to 'M' (Month end) and sum the values
monthly_data = df.resample('M', on='date')[['age_0_5', 'age_5_17', 'age_18_greater']].sum()

# --- STEP 2: PLOT THE STACKED CHART ---
plt.figure(figsize=(12, 7))

# Define colors: Green for babies (growth), Blue for students, Grey for adults
colors = ['#66c2a5', '#3288bd', '#fee08b']
labels = ['0-5 Years (New Births)', '5-17 Years (School Updates)', '18+ Years (Adults)']

# Plotting
plt.stackplot(monthly_data.index,
              monthly_data['age_0_5'],
              monthly_data['age_5_17'],
              monthly_data['age_18_greater'],
              labels=labels,
              colors=colors,
              alpha=0.8)

# --- STEP 3: BEAUTIFY THE CHART ---
plt.title('Monthly Enrollment Trends by Age Group', fontsize=16, fontweight='bold')
plt.xlabel('Timeline', fontsize=12)
plt.ylabel('Number of Enrollments/Updates', fontsize=12)
plt.legend(loc='upper left', fontsize=11)
plt.grid(True, linestyle='--', alpha=0.3)

# Format the x-axis to show dates nicely
plt.gcf().autofmt_xdate()

plt.show()
```

## Observation: 
### A massive jump in the Blue Layer (5-17 Age Group) and Green Layer starting October 2025.

## The Real-World Cause: 
### On October 1, 2025, UIDAI officially waived the charges for Mandatory Biometric Updates (MBU) for children (specifically those who missed the 5/15-year window).

## The Insight: 
### The system is highly price-sensitive. The moment the 125 fee was waived for children in October 2025, enrollments skyrocketed by 600%. This proves that cost was a barrier for rural families.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# --- STEP 1: CREATE THE MISSING MONTH COLUMN ---
# Convert 'date' to datetime objects first (if not already done)
df['date'] = pd.to_datetime(df['date'], format='%d-%m-%Y')

# Create a new column 'Month_Str' (e.g., "2025-10") for grouping
df['Month_Str'] = df['date'].dt.strftime('%Y-%m')

# --- STEP 2: PREPARE DATA FOR HEATMAP ---
# We focus ONLY on the "age_5_17" column (School Children)
# We filter for dates after Sept 2025 to see the "Spike" period clearly
df_school = df[df['date'] >= '2025-09-01'].copy()

# Pivot the data: Rows = State, Columns = Month, Values = Sum of age_5_17
heatmap_data = df_school.pivot_table(index='state', columns='Month_Str', values='age_5_17', aggfunc='sum')

# --- STEP 3: SORT & CLEAN ---
# Sort states by Total Volume so the biggest players are at the top
heatmap_data['Total'] = heatmap_data.sum(axis=1)
heatmap_data = heatmap_data.sort_values(by='Total', ascending=False).head(15) # Top 15 States
heatmap_data.drop(columns=['Total'], inplace=True) # Remove the total column so it doesn't skew the colors

# --- STEP 4: PLOT ---
plt.figure(figsize=(14, 8))

# Plot heatmap using a Red color scale (Darker Red = More Updates)
sns.heatmap(heatmap_data, cmap='Reds', annot=True, fmt='g', linewidths=.5)

# Formatting
plt.title('State-wise Performance: School Age Updates (5-17 Years)', fontsize=16, fontweight='bold')
plt.xlabel('Timeline', fontsize=12)
plt.ylabel('State', fontsize=12)
plt.xticks(rotation=45) # Rotate month labels for readability

plt.tight_layout()
plt.show()
```
